```python
 #3.Delete and remove,clear method examples of elements of dictionary
dic = {103:'ganesh',104:'surya',102:'satya',105:'pavan', 101:'gautham', 106:'advik'}
#pop - to pop specific item based on key
dic.pop(103)
print(dic)
#popitem - to pop the last item
dic.popitem()
print(dic)
#clear - to clear all contents i.e empties the dic & prints empty dict
dic.clear()
print(dic)
# del - to remove the sepcific item and to delete the entire dict 
dic = {103:'ganesh',104:'surya',102:'satya',105:'pavan', 101:'gautham', 106:'advik'}
del dic[104]
print(dic)
del dic
print(dic) #will cause an error
```

```python
#4.To iterate over dictionaries using for loops
dic = {103:'ganesh',104:'surya',102:'satya',105:'pavan', 101:'gautham', 106:'advik'}
for i,j in dic.items():
    print(i,j)
print("\n")
#Keys
for i in dic:
    print(i,end=" ")
print("\n")
for i in dic.keys():
    print(i,end=" ")
print("\n")  
#Values
for i in dic:
    print(dic[i],end=" ")
print("\n")   
for i in dic.values():
    print(i,end=" ")
```
