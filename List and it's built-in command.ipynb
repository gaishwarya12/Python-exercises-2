```python
""""1.Create a list with all types [ex: li = [22,'sachin',True,78.34,21,(1,2,3),'gautham']]

Find length of the list
Reverse the list items with/without using the in built function
Adding the element at the end of the List
Inserting the element in the list
Updating the list element """
li = [22,'sachin',True,78.34,21,(1,2,3),'gautham']
print("Length of the list:",len(li))
print("Reverse with builtin:",li[::-1])
l=len(li)
l=l-1
print("Reverse w/o using builtin:")
while(l>=0):
    print(li[l],end=" ")
    l=l-1
li.append("aishu")
print("\nAdding the element at the end of the List:",li)
li.insert(4,"hello")
print("Inserting the element in the list:",li)
li[4]="hi"
print("Updating the list element:",li)
```

```python
#3.Remove the digits and if reversed item is same as item print it : liril, li = ['sury2a', 'li3ril', 'kreet7hi']
li = ['sury2a', 'li3ril', 'kreet7hi']
li2=[] 
s=""
for i in li:
    for j in i:
        if not j.isdigit():
            s=s+j
    li2.append(s)
    if s==s[::-1]:
        print(s)
    s=""
```

```python
#4.Concatenate two lists index wise [my, her], [Book, pen],Output [myBook, herpen]
l1=["my", "her"]
l2=["Book", "pen"]
l3=[]
for i,j  in zip(l1,l2):
    l3.append(i+j)
print(l3)
```

```python
#5.Remove empty strings from the list of strings
li=['aishu',' ','priya','diya',' ',' ']
l2=[]
for i in li:
    if ' ' not in i:
        l2.append(i)
print(l2)
```

```python
"""6.Given a Python list, find value 20 in the list, 
and if it is present, replace it with 200. Only update the first occurrence of a value"""
li=[10,20,30,40,20,10]
for i in li:
    if i == 20:
        j=li.index(i)
        li[j]=200
        break
print(li)
```

```python
#7.Example for delete, remove, pop, clear: pass and fail cases
li = [10,5,2,90,70,50,100,300,12,25,28]
#del - to delete entire list or to delete multiple items
del li[7:9]
print(li)
#del li
try:
    del li
    print(li)
except:
    print("List not exists") #throws an error
li = [10,5,2,90,70,50,100,300,12,25,28]
#remove - to remove specific item
while True:
    n=int(input())
    if n in li:
        li.remove(n)
        print(li)
    elif n==0:
        break
    else:
        print("Not in list")
# pop to delete last item or to delete particular item using index
li.pop()
print("Last item deleted:",li)
li.pop(6)
print("After deleting element at index 6:",li)
#clear - to remove all the items in list and prints empty list
li.clear()
print(li)
# pop to delete last item or to delete particular item
```

```python
#8.Sort a list using key in ascending and descending
li=['tiger','rat','cat','ant']
li.sort(key=str,reverse=False)
print("Ascending:",li)
li.sort(key=str,reverse=True)
print("Descending:",li)
```
